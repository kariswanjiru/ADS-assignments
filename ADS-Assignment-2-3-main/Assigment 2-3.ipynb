```
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```
Employees = pd.read_csv('WA_Fn-UseC_-HR-Employee-Attrition.csv')
Employees
```

## clean data

```
Employees[Employees.isna().any(axis=1)]
```

```
Employees.info()
```


## show a breakdown of distance from home by job role and attrition

```
Employees.groupby('JobRole')['JobRole'].count()
```

```
Employees.groupby('Attrition')['Attrition'].count()
```

```
Employees.groupby('DistanceFromHome')['DistanceFromHome'].count()
```

```
Employees.groupby(['DistanceFromHome','Attrition'],as_index=False)['EmployeeNumber'].count().pivot('DistanceFromHome','Attrition','EmployeeNumber').fillna(0)
```

```
Employees.groupby(['JobRole','Attrition'],as_index=False)['EmployeeNumber'].count().pivot('JobRole','Attrition','EmployeeNumber').fillna(0)
```

```
Employees.groupby(['JobRole','DistanceFromHome'],as_index=False)['EmployeeNumber'].count().pivot('JobRole','DistanceFromHome','EmployeeNumber').fillna(0)
```

```
sns.relplot(x='EmployeeNumber',y='JobRole',data=Employees,col='DistanceFromHome',hue='Attrition',col_wrap=3)
```

## Compare average moonthly income by education and attrition

```
y = Employees.groupby(['Education'],as_index=False)['MonthlyIncome'].mean().fillna(0)
y
```

```
Employees.groupby(['Education','Attrition'],as_index=False)['EmployeeNumber'].count().pivot('Education','Attrition','EmployeeNumber').fillna(0)
```

```
sns.relplot(x='EmployeeNumber',y='MonthlyIncome',data=Employees,hue='Attrition',col='Education',col_wrap=3)
```

```
plt.hist(y,bins=10)
plt.show()
```
