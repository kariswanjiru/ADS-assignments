Using Plotly express as px, source the data using the command px.data.gapminder() and save it
into a dataframe. Based on the data, map the following information with the appropriate
geographical plot. Map out:

```python
import pandas as pd
import plotly.express as px
import plotly.io as pio
pio.renderers.default = "iframe"
Map = px.data.gapminder()
Map
```

### Question 1

The mean life expectancy of all the Asian countries according to the data

```python
Asia = Map[Map.continent=='Asia']
LifeExpMean = Asia.groupby('country')['lifeExp'].mean()
Asia_LifeExpMean= LifeExpMean.to_frame()
ASIA = pd.merge(Asia,Asia_LifeExpMean,on=['country'])
ASIA.rename(columns={'lifeExp_y': 'lifeExpMean'}, inplace=True)
ASIA
#lifeExp_y is the mean of the country
```

```python
fig =  px.choropleth(ASIA, locations="iso_alpha", color="lifeExpMean", hover_name="country",
                projection="natural earth")
fig
```

### Question 2

Deviation in GDP of each country in Europe and South America.

```python
Europe = Map[Map.continent=='Europe']
SouthAmerica = Map[Map.continent=='Americas']
EuropeAndSouthAmerica=pd.concat([Europe,SouthAmerica])
GDPdeviation=EuropeAndSouthAmerica.groupby('country')['gdpPercap'].std()
EuropeAndSouthAmerica_GDPdeviation=GDPdeviation.to_frame()
EuroAndAme = pd.merge(EuropeAndSouthAmerica,EuropeAndSouthAmerica_GDPdeviation,on=['country'])
EuroAndAme.rename(columns={'gdpPercap_y': 'gdpPercapDeviation'}, inplace=True)
EuroAndAme.head(13)
```

```python
fig =  px.choropleth(EuroAndAme, locations="iso_alpha", color="gdpPercapDeviation", hover_name="country",
                projection="natural earth")
fig
```
